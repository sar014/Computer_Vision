```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

print("Ultralytics YOLOv8 ready")

# Load YOLOv8 nano (fastest, best for Colab)
model = YOLO("yolov8n.pt")

print("Model loaded âœ“")
```

```python
video_path = "/content/yolo.mp4"   # change if needed
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Cannot open video file")
    raise SystemExit

print("Starting near real-time detection...")
```

```python
frame_id = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Video finished.")
        break

    frame_id += 1

    # YOLO inference (people + cars only)
    results = model(
        frame,
        classes=[0, 2],   # 0=person, 2=car
        imgsz=320,
        conf=0.4,
        max_det=40,
        verbose=False
    )

    annotated_frame = results[0].plot()

    # Show every 5th frame to avoid flooding Colab output
    if frame_id % 5 == 0:
        cv2_imshow(annotated_frame)
```
